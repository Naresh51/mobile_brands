```python
import re
```

```python
def extract_ram(About):
    """Extract the RAM from the about."""
    match = re.search(r'RAM(\d+)',About)
    if match:
        return int(match.group(1))
    return None

def extract_system(About):
    """Extract the operating system from the About."""
    match = re.search(r'System(.*?)Processor', About)
    if match:
        return match.group(1).strip()
    return None

def extract_processor(About):
    """Extract the processor type from the About."""
    match = re.search(r'Processor (.*?) ', About)
    if match:
        return match.group(1).strip()
    return None
def extract_storage(About):
    """Extract the storage capacity from the about."""
    match = re.search(r'Storage(\d+)', About)
    if match:
        return int(match.group(1))
    return None

# Apply the functions to the 'About' column
df1['RAM'] = df1['About'].apply(extract_ram)
df1['System'] = df1['About'].apply(extract_system)
df1['Processor'] = df1['About'].apply(extract_processor)
df1['Storage'] = df1['About'].apply(extract_storage)

# Display the first few rows of the DataFrame
df1.head()
```

```python

def clean_system(system):
    """Clean the system values."""
    if pd.isnull(system):
        return None
    if 'android' in system.lower():
        return 'Android'
    if 'ios' in system.lower():
        return 'iOS'
    if 'tizen' in system.lower():
        return 'Tizen'
    return system

# Apply the function to the 'System' column
df1['System'] = df1['System'].apply(clean_system)
# Check the unique values for 'System'
unique_systems_clean = df1['System'].unique()

unique_systems_clean
```

```python
def clean_processor(processor):
    """Clean the processor values."""
    if pd.isnull(processor):
        return None
    if 'mediatek' in processor.lower():
        return 'MediaTek'
    if 'qualcomm' in processor.lower():
        return 'Qualcomm'
    if 'apple' in processor.lower():
        return 'Apple'
    return 'Other'

# Apply the function to the 'Processor' column
df1['Processor'] = df1['Processor'].apply(clean_processor)
# Check the unique values for 'Processor'
unique_processors_clean = df1['Processor'].unique()

unique_processors_clean
```
